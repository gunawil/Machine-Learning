```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../database/Student_performance_data _.csv')
df.describe().T
```

```python
df_viz = df.copy()

gender_label = {0: 'Male', 1: 'Female'}
ethnic_label = {0: 'Caucasian', 1: 'African American', 2: 'Asian', 3: 'Other'}
parent_edu_label = {0: 'None', 1: 'High School', 2: 'Some College', 3: 'Bachelor', 4: 'Higher'}
parent_involve_label = {0: 'None', 1: 'Low', 2: 'Moderate', 3: 'High', 4: 'Very High'}
tutor_label = {0: 'No', 1: 'Yes'}
extra_curr_label = {0: 'No', 1: 'Yes'}
sport_label = {0: 'No', 1: 'Yes'}
music_label = {0: 'No', 1: 'Yes'}
volunteer_label = {0: 'No', 1: 'Yes'}
grade_label = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'F'}

df_viz['Gender'] = df_viz['Gender'].map(gender_label)
df_viz['Ethnicity'] = df_viz['Ethnicity'].map(ethnic_label)
df_viz['ParentalEducation'] = df_viz['ParentalEducation'].map(parent_edu_label)
df_viz['Tutoring'] = df_viz['Tutoring'].map(tutor_label)
df_viz['ParentalSupport'] = df_viz['ParentalSupport'].map(parent_involve_label)
df_viz['Extracurricular'] = df_viz['Extracurricular'].map(extra_curr_label)
df_viz['Sports'] = df_viz['Sports'].map(sport_label)
df_viz['Music'] = df_viz['Music'].map(music_label)
df_viz['Volunteering'] = df_viz['Volunteering'].map(volunteer_label)
df_viz['GradeClass'] = df_viz['GradeClass'].map(grade_label)

df_viz.drop(columns=['StudentID', 'GPA'], inplace=True)

bins_study = [0, 5, 10, 15, 20]
labels_study = ['Low', 'Medium', 'High', 'Very High']
df_viz['StudyTimeCategory'] = pd.cut(df_viz['StudyTimeWeekly'], bins=bins_study, labels=labels_study, right=False)

bins_absence = [0, 3, 7, 15, 30]
labels_absence = ['Very Rare', 'Rare', 'Often', 'Very Often']
df_viz['AbsencesCategory'] = pd.cut(df_viz['Absences'], bins=bins_absence, labels=labels_absence, right=False)

df_viz.head(10)
```

```python
df_viz[df_viz.select_dtypes(include='object').columns] = df_viz.select_dtypes(include='object').apply(lambda x:x.astype('category'))

col_category = df_viz.select_dtypes(include='category').columns.tolist()
col_numeric = df_viz.select_dtypes(exclude='category').columns.tolist()

print(f"\n {col_category}")
print(f"\n {col_numeric}")
```

```python
def category_univariat_plot(df, col_category):
    n_cols = 4
    n_rows = (len(col_category) + n_cols - 1) // n_cols

    plt.figure(figsize=(25, 5 * n_rows))
    
    for index, category in enumerate(col_category):
        plt.subplot(n_rows, n_cols, index + 1)
        # sns.countplot(x=category, data=df)
        data = df[category].value_counts()
        plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis', len(data)))
        plt.title(f"Distribution of {category}")

    plt.tight_layout
    plt.show()

category_univariat_plot(df_viz, col_category)
```

```python
def numeric_univariat_plot(df, col_numeric):
    n_cols = 3
    n_rows = (len(col_numeric) + 1) // n_cols

    plt.figure(figsize=(15, 5 * n_rows))
    
    for index, numeric in enumerate(col_numeric):
        plt.subplot(n_rows, n_cols, index + 1)
        sns.histplot(df[numeric], kde=True)
        plt.title(f"Distribution of {numeric}")
        plt.xlabel(numeric)
        plt.ylabel("Count")

    plt.tight_layout()
    plt.show()


numeric_univariat_plot(df, col_numeric)
```

```python
if 'GradeClass' in col_category:
    col_target = 'GradeClass'
    col_category.remove('GradeClass')
    

def category_bivariat_plot(df, col_category, col_target):
    n_cols = 4
    n_rows = (len(col_category) + n_cols - 1) // n_cols

    plt.figure(figsize=(20, 5 * n_rows))
    cat_order = ['A','B','C','D','F']
    for index, category in enumerate(col_category):
        plt.subplot(n_rows, n_cols, index + 1)
        sns.countplot(x=category, hue=col_target, data=df, palette='viridis', hue_order=cat_order)
        plt.title(f"Distribution of {category} by {col_target}")
        plt.xlabel("")
        plt.xticks(rotation=10)
        plt.ylabel("")

    plt.legend(title=col_target)
    plt.tight_layout
    plt.show()

category_bivariat_plot(df_viz, col_category, col_target)
```

```python
# Fokus pada Dukungan Orang Tua (ParentalSupport)

order = ['None', 'Low', 'Moderate', 'High', 'Very High']

plt.figure(figsize=(10, 6))
sns.countplot(x='ParentalSupport', hue='GradeClass', data=df_viz, palette='viridis', order=order)
plt.title("Distribution of Grades by Parental Support Level")
plt.xlabel("")
plt.ylabel("")
plt.legend(title='Grade Class', loc='upper right')
plt.savefig("Distribution of Grades by Parental Support Level.png")
plt.show()
```
